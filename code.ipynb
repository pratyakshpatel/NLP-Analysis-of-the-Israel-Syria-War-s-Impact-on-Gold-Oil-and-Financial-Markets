```python
import os
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

text_files_directory = r"C:\Users\Pratyaksh\OneDrive - Indian Institute of Science Education and Research Bhopal\projects and assignments\mbfm\General"

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
#we remove these words because they are bound to repeat many times and also dont add anything to out study 
additional_words_to_remove = ['israel', 'hamas', 'palestine', 'finance', 'news', 'economic', 'effect', 'iran', 'iranisrael', 'israelhamas', 'israeliran', 'israelpalestine'
                              'risk', 'middle', 'east', 'national', 'international', 'relation', 'relations', 'index', 'fund', 'war', 
                              'facebook', 'social', 'media', 'government', 'factors', 'policy', 'make', 'thereatened', 'part', 'series', 'market', 'cycle', 'attack',
                              'member', 'members', 'aid', 'measures', 'security', 'dynamic', 'forces', 'conflict', 'gaza', 'mint', 'subscribe', 'official',
                              'people', 'support', 'fight', 'fightting', 'latest', 'border', 'cease', 'fire', 'casualties', 'trend', 'trending',
                              'updates', 'update', 'live', 'except', 'nov', 'today', 'best', 'calculator', 'could', 'view', 'per','world', 'said', 'top', 'india', 'indian', 'program', 'data', 'would', 'year', 'event', 'read'
                               , 'account', 'story', 'january','february', 'march' ,'april', 'may', 'june', 'july', 'august', 'september', 'october', 'november', 'december', 
                               'monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday' ]
                            

def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove non-alphabetic characters
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = text.lower()  # Convert to lowercase
    words = text.split()
    words = [word for word in words if word not in stop_words and len(word) > 2]  # Remove stopwords and short words
    words = [lemmatizer.lemmatize(word) for word in words]  # Lemmatize words
    words = [word for word in words if word not in additional_words_to_remove]  # Remove additional words
    cleaned_text = ' '.join(words)
    return cleaned_text

processed_texts = []

for filename in os.listdir(text_files_directory):
    if filename.endswith('.txt'):
        file_path = os.path.join(text_files_directory, filename)
        
        with open(file_path, 'r', encoding='utf-8') as file:
            file_content = file.read()
            cleaned_text = clean_text(file_content)
            processed_texts.append(cleaned_text)
            print(f"Processed and appended: {filename}")

print(f"Total number of files processed and appended: {len(processed_texts)}")

processed_texts
```

The LDA model is used to find different topic models.


What is LDA ? 
Latent Dirichlet Allocation (LDA) is a probabilistic model commonly used for topic modeling, which helps uncover hidden topics in a collection of documents. It assumes that each document is composed of a mixture of topics, and each topic is a probability distribution over a set of words. LDA uses Dirichlet distributions to control the proportion of topics in each document and the distribution of words within each topic, ensuring sparsity and interpretability. By assigning probabilities to words and topics, LDA can identify patterns and relationships within textual data, making it a valuable tool for organizing and summarizing large text corpora.

```python
import gensim
from gensim.corpora import Dictionary
from gensim.models import LdaModel
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')

tokenized_texts = [word_tokenize(text) for text in processed_texts]

word_dict = Dictionary(tokenized_texts)

bow_corpus = [word_dict.doc2bow(text) for text in tokenized_texts]

topic_range = list(range(1, 16))
num_keywords_per_topic = 15  

lda_models = {}
lda_topics = {}

for num_topics in topic_range:
    lda_models[num_topics] = LdaModel(
        corpus=bow_corpus,
        id2word=word_dict,
        num_topics=num_topics,
        update_every=1,
        chunksize=len(bow_corpus),
        passes=20,
        alpha='auto',
        random_state=42
    )
    
    topics = lda_models[num_topics].show_topics(num_topics=num_topics, 
                                                num_words=num_keywords_per_topic,
                                                formatted=False)
    
    lda_topics[num_topics] = [[word[0] for word in topic[1]] for topic in topics]

for num, topics in lda_topics.items():
    print(f"Model with {num} topics:")
    for idx, topic in enumerate(topics):
        print(f"  Topic {idx + 1}: {topic}")
```

Now we will compute the coherence score for each topic. The coherence score is a metric used to evaluate the quality of topics generated by topic modeling techniques like LDA. It measures the semantic similarity between the most significant words in a topic by analyzing how often they appear together in the text corpus. A high coherence score indicates that the words in a topic are meaningfully related, making the topic more interpretable and relevant to the underlying data. This is important because it ensures that the generated topics are not just mathematical abstractions but also align with human understanding, aiding in better insights and actionable conclusions from the modeled data.

```python
from gensim.models import CoherenceModel

topic_counts = list(range(1, 16))  # Topics from 1 to 15

lda_coherence_scores = {}

for count in topic_counts:
    coherence_model = CoherenceModel(
        model=lda_models[count],      
        texts=tokenized_texts,         
        dictionary=word_dict,          
        coherence='c_v'                
    )
    coherence_score = coherence_model.get_coherence()
    lda_coherence_scores[count] = coherence_score

for count, score in lda_coherence_scores.items():
    print(f"Coherence Score for {count} topics: {score:.4f}")
```

Mean overlap is a metric used to evaluate the distinctiveness of topics generated by a topic modeling algorithm. It measures the average proportion of shared words between pairs of topics, with lower values indicating better topic separation. A high mean overlap suggests that topics are not well-differentiated, leading to redundancy and reduced interpretability. Ensuring a low mean overlap is important for generating a diverse set of topics that capture distinct themes within the dataset, improving the overall quality and usability of the topic modeling results.

```python
from gensim.models import CoherenceModel
import numpy as np

def jaccard_similarity(set1, set2):
    intersection = len(set(set1) & set(set2))
    union = len(set(set1) | set(set2))
    return intersection / union if union != 0 else 0

topic_counts = list(range(1, 16))

lda_overlap_scores = {}

for i in range(len(topic_counts) - 1):
    overlap_similarities = []
    topics_current_model = lda_topics[topic_counts[i]]
    topics_next_model = lda_topics[topic_counts[i + 1]]
    
    for topic1 in topics_current_model:
        similarities = [jaccard_similarity(topic1, topic2) for topic2 in topics_next_model]
        overlap_similarities.append(similarities)
    
    lda_overlap_scores[topic_counts[i]] = overlap_similarities

mean_overlap_scores = {num: np.mean(overlaps) for num, overlaps in lda_overlap_scores.items()}

for num, score in mean_overlap_scores.items():
    print(f"Mean Topic Overlap for {num} topics: {score:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def jaccard_similarity(set1, set2):
    intersection = len(set(set1) & set(set2))
    union = len(set(set1) | set(set2))
    return intersection / union if union != 0 else 0

num_topics = list(range(1, 16)) 

LDA_overlap = {}
for i in range(len(num_topics) - 1):
    overlap_sims = []
    for topic1 in lda_topics[num_topics[i]]:
        sims = [jaccard_similarity(topic1, topic2) for topic2 in lda_topics[num_topics[i + 1]]]
        overlap_sims.append(sims) 
    LDA_overlap[num_topics[i]] = overlap_sims 

mean_overlap = {}
for num, overlap in LDA_overlap.items():
    flat_overlap = [score for sublist in overlap for score in sublist]
    mean_overlap[num] = np.mean(flat_overlap) if flat_overlap else 0  

coherence_scores = [lda_coherence_scores[num] for num in num_topics]
overlap_scores = [mean_overlap.get(num, 0) for num in num_topics[:-1]]

max_coherence = max(coherence_scores)
min_overlap = min(overlap_scores)
max_coherence_index = coherence_scores.index(max_coherence)
min_overlap_index = overlap_scores.index(min_overlap)

plt.figure(figsize=(10, 6))
plt.plot(num_topics, coherence_scores, marker='o', label='Coherence Score', color='blue')
plt.scatter(num_topics[max_coherence_index], max_coherence, color='red', s=100, label='Max Coherence', zorder=5)
plt.plot(num_topics[:-1], overlap_scores, marker='o', label='Mean Overlap', color='orange')
plt.scatter(num_topics[min_overlap_index], min_overlap, color='green', s=100, label='Min Overlap', zorder=5)

plt.title('Coherence Scores and Topic Overlap vs. Number of Topics')
plt.xlabel('Number of Topics')
plt.ylabel('Scores')
plt.xticks(num_topics)
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
